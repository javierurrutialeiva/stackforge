```python
basePath = "/home/tnguser/sims.TNG/TNG100-3/output/"
redshifts = [0, 0.5, 1]
haloIDs = [i for i in range(30)]
R = np.linspace(0, 15000, 100)
for z in redshifts:
    snap = stackforge.load_snap(basePath, redshift = z)
    s = stackforge.stack(sim = "TNG", None, haloIDs, True, redshift=z, snap = snap)
    s.compute_1h2hprofiles(R, comoving = True, remove_h = True)
    s.save()
```

```python
R = np.linspace(0, 15000, 100)
h = 0.67
redshifts = ["0.0", "0.5", "1.0"]
files = [f"profiles_z={z}.h5" for z in redshifts]
hfiles = [h5py.File(f) for f in files]
p1h_profiles = np.array([h["profiles1h"][:,0] for h in hfiles])
p2h_profiles = np.array([h["profiles2h"][:,0] for h in hfiles])
mean = np.array([float(np.array(h["cmean_ne"])) for h in hfiles])
ptotal = p1h_profiles + p2h_profiles
ptotal = ptotal - mean[:,None,None]/h**3
pmean = np.mean(ptotal, axis = 1)
p1halo = np.mean(p1h_profiles, axis = 1)
p2halo = pmean - p1halo
fig, ax = plt.subplots(figsize = (12, 8))
Rbins = (R[1::] + R[-1::])/2
for i in range(len(pmean)):
    line, = ax.loglog(Rbins, pmean[i], lw = 2, alpha = 0.8, label = "$z = %.2f$" % float(redshifts[i]))
    ax.loglog(Rbins, p1halo[i], lw = 2, alpha = 0.8, color = line.get_color(), ls = "--")
    ax.loglog(Rbins, p2halo[i], lw = 2, alpha = 0.8, color = line.get_color(), ls = "dotted")
ax.plot([],[], label = "1halo", ls = "--", color = "black")
ax.plot([],[], label = "2halo", ls = "dotted", color = "black")
ax.plot([],[], label = "total", color = "black")
ax.set_ylim(bottom=1e-2)
ax.set(xlabel = "R (ckpc)", ylabel = "electron number density")
ax.legend(frameon = False)
fig.savefig("profiles_vs_redshift.png")
```

```python
Masses = (np.array([h["Masses"][:]/h["h"] for h in hfiles]) * 1e10).flatten()
fptotal = np.reshape(ptotal, (len(ptotal)*len(ptotal[0]),-1))
fp1h = np.reshape(p1h_profiles, (len(ptotal)*len(ptotal[0]),-1))
Mbins = np.logspace(13.5, 15, 4)
fig, ax = plt.subplots(figsize = (12, 8))

for i in range(len(Mbins) - 1):
    Mmin = Mbins[i]
    Mmax = Mbins[i + 1]
    mask = np.where((Masses >= Mmin) & (Masses < Mmax))
    pt = fptotal[mask] - np.mean(mean)/h**3
    pmean = np.mean(pt, axis = 0)
    p1h = fp1h[mask]
    p2h = pt - p1h
    pmean = np.mean(pt, axis = 0) 
    p1hmean = np.mean(p1h, axis = 0)
    p2hmean = np.mean(p2h, axis = 0)
    line, = ax.loglog(Rbins, pmean, lw = 2, alpha = 0.8, label = "$M \in [%.2f,%.2f]$" % (np.log10(Mmin), np.log10(Mmax)))
    ax.loglog(Rbins, p1hmean, lw = 2, alpha = 0.8, color = line.get_color(), ls = "--")
    ax.loglog(Rbins, p2hmean, lw = 2, alpha = 0.8, color = line.get_color(), ls = "dotted")
ax.plot([],[], label = "1halo", ls = "--", color = "black")
ax.plot([],[], label = "2halo", ls = "dotted", color = "black")
ax.plot([],[], label = "total", color = "black")
ax.legend(frameon = False)
ax.set_ylim(bottom=1e-2)
ax.set(xlabel = "R (ckpc)", ylabel = "electron number density $[m^{-3}]$")
fig.savefig("profiles_vs_mass.png")
```

```python
gass_masses = []
total_masses = []
real_masses = []
for i in range(len(s.halos)):
    h = s.halos[i]
    E = s.halos[i].profiles3D[-1]
    ne = s.halos[i].profiles3D[0]
    R = s.halos[i].R_centers
    density = np.nan_to_num(const.m_p.value*ne/(0.76*E))
    R_m = R*u.kpc.to(u.m)
    gas_mass = np.trapz(density*R_m**2*4*np.pi, x = R_m)
    gas_mass = gas_mass*u.kg.to(u.Msun)
    total_mass = gas_mass/0.15
    total_masses.append(total_mass)
    masses = h.gas["Masses"]*1e10/h.h
    total_gas_mass = np.sum(masses)
    total_mass = total_gas_mass/0.15
    total_mass = h.Mass*1e10/h.h
    real_masses.append(total_mass)
total_masses = np.array(total_masses)
real_masses = np.array(real_masses)
```

```python
from matplotlib.gridspec import GridSpec

fig = plt.figure(figsize=(8, 5))
gs = GridSpec(2, 1, height_ratios=[4, 1], hspace=0.05, figure=fig)

ax_main = fig.add_subplot(gs[0])
ax_ratio = fig.add_subplot(gs[1], sharex=ax_main)

ax_main.scatter(real_masses, total_masses, s = 50, alpha = 0.5, color = "darkred", edgecolor = "black")
ax_main.plot(real_masses, real_masses, ls = "--", label = "1:1", color = "black", lw = 3, alpha = 0.5)
ax_main.legend(frameon = False)
ax_main.set(ylabel = "Recovered total mass", xscale = "log", yscale = "log")
ratio = (total_masses - real_masses)/real_masses
ax_ratio.scatter(real_masses, ratio, s = 50, alpha = 0.5, color = "darkred", edgecolor = "black")
ax_ratio.set(xlabel = "Real Mass", ylabel = "ratio")
ax_ratio.axhline(0)
ax_ratio.set_ylim((-1.5*np.max(np.abs(ratio)), 1.5*np.max(np.abs(ratio))))
fig.savefig("mass_test.png")
```
